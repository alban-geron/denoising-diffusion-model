```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

# --- Configuration ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
LEARNING_RATE = 5e-5
BATCH_SIZE = 64
IMAGE_SIZE = 28
CHANNELS = 1
Z_DIM = 100  # Dimension du vecteur de bruit latent
NUM_EPOCHS = 20 # Augmenter pour de meilleurs résultats
FEATURES_CRITIC = 64
FEATURES_GEN = 64
CRITIC_ITERATIONS = 5 # Entraîner le critique 5 fois pour 1 pas du générateur
WEIGHT_CLIP = 0.01 # Borne pour le clipping des poids (Spécifique au WGAN)

# --- Préparation des Données (MNIST) ---
transforms = transforms.Compose([
    transforms.Resize(IMAGE_SIZE),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5]), # Normalisation entre -1 et 1
])

dataset = torchvision.datasets.MNIST(root="dataset/", train=True, transform=transforms, download=True)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

# --- Définition des Modèles ---

class Critic(nn.Module):
    def __init__(self, channels_img, features_d):
        super(Critic, self).__init__()
        self.critic = nn.Sequential(
            # Input: N x channels_img x 28 x 28
            nn.Conv2d(channels_img, features_d, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),

            # Input: 14 x 14
            nn.Conv2d(features_d, features_d * 2, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(features_d * 2),
            nn.LeakyReLU(0.2),

            # Input: 7 x 7
            nn.Conv2d(features_d * 2, features_d * 4, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(features_d * 4),
            nn.LeakyReLU(0.2),

            # Input: 3 x 3 (Output final convolutif vers une seule valeur)
            nn.Conv2d(features_d * 4, 1, kernel_size=4, stride=2, padding=1), # Changed padding from 0 to 1
            # Note importante : PAS de Sigmoid à la fin pour un WGAN !
        )

    def forward(self, x):
        return self.critic(x)

class Generator(nn.Module):
    def __init__(self, channels_noise, channels_img, features_g):
        super(Generator, self).__init__()
        self.net = nn.Sequential(
            # Input: N x channels_noise x 1 x 1
            nn.ConvTranspose2d(channels_noise, features_g * 16, kernel_size=4, stride=1, padding=0),
            nn.BatchNorm2d(features_g * 16),
            nn.ReLU(),

            # 4x4
            nn.ConvTranspose2d(features_g * 16, features_g * 8, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(features_g * 8),
            nn.ReLU(),

            # 8x8
            nn.ConvTranspose2d(features_g * 8, features_g * 4, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(features_g * 4),
            nn.ReLU(),

            # 16x16 -> Il faut ajuster pour arriver à 28x28 (MNIST est tricky avec les puissances de 2)
            # Ici on fait simple : on vise 32x32 et on resize ou on utilise des padding spécifiques.
            # Pour simplifier ce code, on va utiliser une architecture adaptée à 28x28:
            nn.ConvTranspose2d(features_g * 4, channels_img, kernel_size=4, stride=2, padding=3),
            nn.Tanh() # Output: N x channels_img x 28 x 28. Tanh pour sortie entre -1 et 1
        )

    def forward(self, x):
        return self.net(x)

# --- Initialisation ---
gen = Generator(Z_DIM, CHANNELS, FEATURES_GEN).to(device)
critic = Critic(CHANNELS, FEATURES_CRITIC).to(device)

# Initialisation des poids (Recommandé pour le WGAN)
def initialize_weights(model):
    for m in model.modules():
        if isinstance(m, (nn.Conv2d, nn.ConvTranspose2d, nn.BatchNorm2d)):
            nn.init.normal_(m.weight.data, 0.0, 0.02)

gen.apply(initialize_weights)
critic.apply(initialize_weights)

# Optimiseurs (RMSprop est recommandé pour le WGAN original au lieu d'Adam)
opt_gen = optim.RMSprop(gen.parameters(), lr=LEARNING_RATE)
opt_critic = optim.RMSprop(critic.parameters(), lr=LEARNING_RATE)

fixed_noise = torch.randn(32, Z_DIM, 1, 1).to(device)

# --- Boucle d'Entraînement ---
print("Début de l'entraînement...")

for epoch in range(NUM_EPOCHS):
    for batch_idx, (real, _) in enumerate(loader):
        real = real.to(device)
        cur_batch_size = real.shape[0]

        # 1. Entraînement du Critique (Discriminator)
        for _ in range(CRITIC_ITERATIONS):
            noise = torch.randn(cur_batch_size, Z_DIM, 1, 1).to(device)
            fake = gen(noise)

            critic_real = critic(real).reshape(-1)
            critic_fake = critic(fake).reshape(-1)

            # Fonction de perte WGAN : -(E[D(x)] - E[D(G(z))])
            # On veut maximiser l'écart, donc minimiser le négatif
            loss_critic = -(torch.mean(critic_real) - torch.mean(critic_fake))

            critic.zero_grad()
            loss_critic.backward(retain_graph=True)
            opt_critic.step()

            # CLIP DES POIDS (Spécifique WGAN)
            for p in critic.parameters():
                p.data.clamp_(-WEIGHT_CLIP, WEIGHT_CLIP)

        # 2. Entraînement du Générateur
        # On n'entraîne le générateur qu'une fois tous les n_critic pas
        gen_fake = critic(fake).reshape(-1)
        loss_gen = -torch.mean(gen_fake) # On veut maximiser D(G(z))

        gen.zero_grad()
        loss_gen.backward()
        opt_gen.step()

        # Affichage
        if batch_idx % 100 == 0 and batch_idx > 0:
            print(f"Epoch [{epoch}/{NUM_EPOCHS}] Batch {batch_idx}/{len(loader)} \
                  Loss D: {loss_critic:.4f}, loss G: {loss_gen:.4f}")

    # Visualisation à chaque epoch
    with torch.no_grad():
        fake = gen(fixed_noise)
        img_grid_fake = torchvision.utils.make_grid(fake[:32], normalize=True)

        plt.imshow(img_grid_fake.permute(1, 2, 0).cpu().numpy())
        plt.title(f"Images Générées - Epoch {epoch}")
        plt.axis("off")
        plt.show()
```
